```python
def plot_loss(neuron_type):
    plt.plot(losses[0], label='{0} 30 neurons'.format(neuron_type))
    plt.legend(bbox_to_anchor=(0.8, 1), loc=2, borderaxespad=0.)
    plt.xlabel('epochs')
    plt.ylabel('Loss value')
    plt.axis([0, 300, 0, 0.03])
    plt.savefig('{0} learning loss.png'.format(neuron_type))
    plt.show()
```

```python
def plot_predictions(neuron_type, frames=30):
    fig = plt.figure(dpi=100)
    ax = plt.axes(xlim=(-4, 4), ylim=(-1, 1))
    ax.grid()

    ttl = ax.text(0.5, 1.05, '', transform = ax.transAxes, va='bottom')
    ax.plot(X, y, label='Function y = sin(x)')
    line, = ax.plot([], [], lw=2)
    line.set_label('Model with 30 {0} neurons'.format(neuron_type))

    def init():
        ttl.set_text('')
        line.set_data(X, learn_predictions[0][0])
        return line, ttl

    def animate(i):
        ttl.set_text("Epoch {0}, loss {1}".format(i*10, losses[0][i*10]))
        x = X
        y = learn_predictions[0][i]
        line.set_data(x, y)
        return line, ttl

    ax.legend()

    anim = animation.FuncAnimation(fig, animate, init_func=init,
                                   frames=frames, interval=150, blit=True)

    anim.save('Model_with_{0}_{1}_neurons.gif'.format(30, neuron_type), 
              dpi=100, writer='imagemagick')
```

```python
def fit_model(neuron_type):
    
    learn_predictions = [] #set of predictions by every 10 epoch while learning
    losses = []            #set of loss values while learning

    model = Sequential()

    model.add(Dense(30, input_dim = 1, activation=neuron_type, kernel_initializer='he_normal'))
    model.add(Dense(1, activation='linear'))

    model.compile(loss='mse', optimizer=Optimizer) 

    loss = []
    predicted = []

    for i in range(30):
        print('Model with {0} {2} neurons, Epoch {1}.'.format(30, i*10, neuron_type))
        hist = model.fit(x_train, y_train, batch_size=1, epochs=10, verbose=2)

        predicted.append(model.predict(X).reshape(X.shape[0]).tolist())

        loss.extend(hist.history['loss'])

    losses.append(loss)
    learn_predictions.append(predicted)
    
    model.save('sin_1_layer_30_{0}_neurons.h5'.format(neuron_type))
    
    return losses, learn_predictions
```

```python
def additional_fitting(model_name):
    
    learn_predictions = [] #set of predictions by every 10 epoch while learning
    losses = []            #set of loss values while learning
    
    model = load_model(model_name)
    
    loss = []
    predicted = []

    for i in range(20):
        print('Model {0}, Epoch {1}.'.format(model_name, i*10))
        hist = model.fit(x_train, y_train, batch_size=1, epochs=10, verbose=2)

        predicted.append(model.predict(X).reshape(X.shape[0]).tolist())

        loss.extend(hist.history['loss'])

    losses.append(loss)
    learn_predictions.append(predicted)
    
    model.save('Retrained_{0}'.format(model_name))
    
    return losses, learn_predictions
```

### Build and fit model
Model with 2 layers:
* 1 layer: 30 neurons, **selu** activation;
* 1 neuron, activation linear.

Epochs number: 300
Optimizer: Adam
Batch_size: 1

$ SELU(x,\alpha)\begin{cases} x & if\ x>0 \\ \alpha e^{x}-\alpha & if\ x \leq 0 \end{cases} $

This how relu and selu looks like.

```python
def relu(x):
    return max(0,x)

def selu(x, alpha=1):
    return x if x>0 else alpha*math.exp(x)-alpha

plt.plot(np.linspace(-5,1,num=300), list(map(relu, np.linspace(-5,1,num=300))),label='relu')
plt.plot(np.linspace(-5,1,num=300), list(map(selu, np.linspace(-5,1,num=300))),label='selu')
plt.grid()
plt.legend()
plt.show()
```

```python
losses, learn_predictions = fit_model('selu')
```

Plot loss values

```python
plot_loss('selu')
```

```python
plot_predictions('selu')
```

### Build and fit model

Sigmoid function is:
$ \sigma(x)= \dfrac {1}{1+e^{-x}} $

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))

plt.plot(np.linspace(-8,8,num=300),
         list(map(sigmoid, np.linspace(-8,8,num=300))),
         label='sigmoid')

plt.grid()
plt.legend()
plt.show()
```

### Build and fit model
Model with 2 layers:
* 1 layer: 30 neurons, **softplus** activation;
* 1 neuron, activation linear.

Epochs number: 300
Optimizer: Adam
Batch_size: 1

Softplus function is:
$ softplus(x)= \ln (1+e^{x}) $

Softplus is a smooth version of relu.

```python
def softplus(x):
    return math.log(1+math.exp(x))

arg = np.linspace(-4,4,num=300)

plt.plot(arg,
         list(map(softplus, arg)),
         linewidth=4.0,
         label='softplus')

plt.plot(arg,
         list(map(relu, arg)),
         label='relu')

plt.plot(arg,
         list(map(selu, arg)),
         'r',
         label='selu')

plt.grid()
plt.legend()
plt.show()
```

```python
losses, learn_predictions = fit_model('softplus')
```

```python
plot_loss('softplus')
```

```python
plot_predictions('softplus')
```

### Build and fit model
Model with 2 layers:
* 1 layer: 30 neurons, **elu** activation;
* 1 neuron, activation linear.

Epochs number: 300
Optimizer: Adam
Batch_size: 1

ELU function is:

$ ELU(x,\alpha)\begin{cases} x & if\ x>0 \\ \alpha(e^{x}-1) & if\ x \leq 0 \end{cases} $

There's one trouble with relu: while learning some neurons can die, so selu and elu neurons was made to make it better.

```python
def elu(x, alpha=1):
    return x if x>0 else alpha*(math.exp(x)-1)

arg = np.linspace(-4,4,num=300)

plt.plot(arg,
         list(map(elu, arg)), 'r',
         linewidth=2.0,
         label='elu, alpha=1')

plt.plot(arg,
         list(map(relu, arg)),'b',
         label='relu')


plt.grid()
plt.legend()
plt.show()
```

```python
losses, learn_predictions = fit_model('elu')
```

```python
plot_loss('elu')
```

```python
plot_predictions('elu')
```

It's  doesn't look's nice, cause after 300 epochs, models with 1 hidden layer with 30 neurons can't normal predict the original function. Now, I'll try to find, which model has best score on MSE:

```python
scores = []
for file in os.listdir():
    if file.endswith(".h5"):
        
        model = load_model(file)
        scores.append([file, model.evaluate(X,y)])
```

```python
for score in scores:
    print("Model: {0},     score: {1}".format(score[0][15:-11], score[1]))

print("\n Best score have model:")
print(scores[[score[1] for score in scores].index(min([score[1] for score in scores]))][0][15:-11])
```

# Part 3: Maybe is 300 epochs doesn't enough?
So, now for every model i'll try to increment learn steps on 200.

```python
out = {}
for file in os.listdir():
    if file.endswith(".h5"):
        losses, learn_predictions = additional_fitting(file)
        out[file] = {
            'losses': losses,
            'learn_predictions': learn_predictions
        }
```

Loss by every epochs

```python
for model in out:
    plt.plot(out[model]['losses'][0], label='{0}'.format(model))
    plt.legend(bbox_to_anchor=(0.8, 1), loc=2, borderaxespad=0.)
    plt.xlabel('epochs')
    plt.ylabel('Loss value')
    plt.axis([0, 200, 0, 0.01])
    plt.savefig('{0}.png'.format(model))
    plt.grid()
    plt.show()
```

```python
for model in out:
    plt.plot(out[model]['losses'][0], '.-', label='{0}'.format(model))
    
plt.legend(bbox_to_anchor=(0.8, 1), loc=2, borderaxespad=0.)
plt.xlabel('epochs')
plt.ylabel('Loss value')
plt.axis([0, 100, 0, 0.01])
plt.savefig('{0}.png'.format('Loss'))
plt.grid()
plt.show()
```

```python
for model in out:
    fig = plt.figure(dpi=100)
    ax = plt.axes(xlim=(-4, 4), ylim=(-1, 1))
    ax.grid()

    ttl = ax.text(0.5, 1.05, '', transform = ax.transAxes, va='bottom')
    ax.plot(X, y, label='Function y = sin(x)')
    line, = ax.plot([], [], lw=2)
    line.set_label('{0}'.format(model))

    def init():
        ttl.set_text('')
        line.set_data(X, out[model]['learn_predictions'][0][0])
        return line, ttl

    def animate(i):
        ttl.set_text("Epoch {0}".format(i*10+300))
        x = X
        y = out[model]['learn_predictions'][0][i]
        line.set_data(x, y)
        return line, ttl

    ax.legend()

    anim = animation.FuncAnimation(fig, animate, init_func=init,
                                   frames=20, interval=150, blit=True)

    anim.save('{0}.gif'.format(model), 
              dpi=100, writer='imagemagick')
```
